# K-means

```python
#Note that we have to specify no of clusters k that the algorithm must find. 
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
x,y=make_blobs(n_samples=500,centers=5)
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
k=5
kmeans=KMeans(n_clusters=k,random_state=42,n_init=1)
y_pred=kmeans.fit_predict(x)
```

```python
y_pred
```

```python
kmeans.labels_
```

```python
y_pred is kmeans.labels_
```

```python
#THe centroids are
kmeans.cluster_centers_
```

```python
plt.scatter(x[:,0],x[:,1])
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='X')
```

```python
#The kmeans clustering does not perform well when the blobs have very different diameters because all it cares about
#assigning an instance to a cluster is the distance to the centroid.
```

```python
#Instead of assigning each instance to a single cluster, which is called the hard clustering, it can be useful to give each
#instance a score per cluster, which is called soft clustering. The score can be the distance between the instance and the
#centroid or a similarity score(or affinity), such as the Gaussian radial basis function.The KMeans class, the transform()
#measures the distance from each instance to every centroid
```

```python
kmeans.transform(x[:3])
```

```python
kmeans.predict(x[:3])
```

```python
#The instances with small distance out of all centers is assigned to that cluster(0,1,2,3,4)
```

```python
#If we have a high dimensional data set , this transformation makes it to get to a k-demensional data set. Hence this is 
#a very efficient non linear dimensionality reduction technique. Hence the name transorm came to this function.
#Alternatively, we can use these distances as extra features to train another model.
```
